```python
df["total score"]= df[["math score",  "reading score",  "writing score"]].sum(axis=1) # yangi total score nomli feature hosil qilib oldik.
df["average score"]= df["total score"]/3 # Tanlangan column namelarning umumiy yiginddisini 3 gabolibb ortacha qiymatini oldik.yangi average nomli featurega qiymat sifatida berdik.
df["passed"]= np.where(df["average score"]>= 50, "Yes", "No" ) # agar ortacha qiymat 50 dan yuqori bolsa yes aks holda no matni chiqadi.
print(df[["total score", "average score", "passed"]].head()) # barchasini jadval korinishda chop etdik
```

```python
test_prep_group = df.groupby("test preparation course")[["math score", "reading score", "writing score"]].mean() # yangi ozgaruvchi hosil qilib uni "test preparation course" fani boyichan(completed va none qiymatlari) guruhlarga ajratdik
print(test_prep_group) #natija test_prep_group fanini oqigan talabalar (completed qilganlar) bu fanni oqimagan talabalar(none) ga nisbatan qolgan fanni ozlashtirishining ortacha qiymati yuqori ekan.

# Jins boyicha guruhlab natijani koramiz
gender_avg = df.groupby("gender")[["math score", "reading score", "writing score"]].mean()
print(gender_avg) #demak ayollar erkaklarga nisbatan matematikani kamroq ozlashtirar ekan va hokazo
```

```python
pass_rate = df["passed"].value_counts(normalize=True)*100 #yangi pass_rate hosil qildik, yuqoriroqda hosil qilganimiz "passedni" ustida dfning static value_counts methodini chaqirib uni normalize ture *100 qilinb natijani foizda korsatishni belgiladik.
print(pass_rate)
```

```python
#Convert scores to Numpy array
scores = df[["math score",  "reading score",  "writing score"]].values # values tabledan numpy arrayga (matritsaga) otkazadi.

# Mean, Std deviation
print("Numpay mean:",np.mean(scores, axis=0))
print("Numpay std:",np.std(scores, axis=0))

#Z-score (normalize)
z_score = (scores - np.mean(scores, axis=0)) /np.std(scores, axis=0)
print(z_score[:5])
```

```python
# nechta talaba 90 baldan yuqori olganini hisoblaymiz.
top_scores = df[df["average score"] >=90]
print("Top scored students above 90:", len(top_scores))

#nechta talaba imtihondan yiqildi?
failed = df[df["passed"]== "No" ]
print("Failed students:", len(failed))
```
