```python
# Import PyTorch and matplotlib
import torch
from torch import nn # nn contains all of PyTorch's building blocks for neural networks
import matplotlib.pyplot as plt

# Check PyTorch version
torch.version
```

```python
# Setup device agnostic code
device = "cuda" if torch.cuda.is_available() else "cpu"
print(f"Using device: {device}")
```

```python
# Create weight and bias
weight = 0.7
bias = 0.3

# Create range values
start = 0
end = 1
step = 0.02

# Create X and y (features and labels)
X = torch.arange(start, end, step).unsqueeze(dim=1) # without unsqueeze, errors will happen later on (shapes within linear layers)
y = weight * X + bias

#Linear regression: y= wx+b;
print("X:", X)
print("\n y:", y)

# Split data
train_split = int(0.8 * len(X)) # 100 data => 80 train | 20 for test
X_train, y_train = X[:train_split], y[:train_split] # 80% datani train uchun ishlatamiz
X_test, y_test = X[train_split:], y[train_split:] #train split uchun ishlatilgan data tugagandan boshlab qolgan qismni test uchun ishlat
len(X_train),len(y_train),len(X_test),len(y_test),
```

```python
def plot_predictions(train_data=X_train,
                     train_labels=y_train,
                     test_data=X_test,
                     test_labels=y_test,
                     predictions=None):
    plt.figure(figsize=(10, 7))

    # Plot training data in blue
    plt.scatter(train_data, train_labels, c='b', s=4, label='Training data')

    # Plot test data in green
    plt.scatter(test_data, test_labels, c='g', s=4, label='Testing data')

    if predictions is not None:
        plt.scatter(test_data, predictions, c='r', s=4, label='Predictions')

    plt.legend(prop={'size': 14})

plot_predictions()
```

```python
import torch
from torch import nn

class LinearRegressionModelV2(nn.Module):
    # Ikki tomonida ikkitadan pastki chiziq bo'lishi shart: __init__
    def __init__(self):
        # Ota klass init funksiyasini chaqiramiz
        super().__init__()
        # Chiziqli qatlam yaratamiz
        self.linear_layer = nn.Linear(in_features=1, out_features=1)

    # Model forward funksiyasi
    def forward(self, x: torch.Tensor) -> torch.Tensor:
        return self.linear_layer(x)

# Tasodifiy sonlarni qulflash
torch.manual_seed(42)

# Model obyekti
model_1 = LinearRegressionModelV2()

# Natijalarni ekranga chiqarish
print("--- Model Strukturasi ---")
print(model_1)

print("\n--- Model Parametrlari (state_dict) ---")
print(model_1.state_dict())
```

```python
# Loss (Xatolikni o'lchash) funksiyasini yaratish
# L1Loss - bu MAE (Mean Absolute Error) ya'ni o'rtacha absolut xatolikni hisoblaydi
loss_fn = nn.L1Loss()

# Optimizatorni (modelni o'rgatuvchi mexanizm) sozlash
# SGD - Stoxastik Gradient Descent algoritmi. Unga model parametrlarini va o'rganish tezligini (lr=0.01) beramiz
optimizer = torch.optim.SGD(params=model_1.parameters(),
                            lr=0.01)
```

```python
# Model hali o'qitilmasdan oldin (dastlabki tasodifiy holatida) test ma'lumotlariga qanday bashorat berishini ko'ramiz
model_1.eval() # Modelni baholash (test) rejimiga o'tkazamiz

# Inference (gradientlarni hisoblamaslik) rejimida kodni bajaramiz (bu xotirani tejaydi va tezlashtiradi)
with torch.inference_mode():
    # Test ma'lumotlarini (X_test) modelga berib dastlabki bashoratlarni olamiz
    y_preds_early = model_1(X_test)
    plot_predictions(predictions=y_preds_early)
    plt.title("Predictions before training( red random guesses)")
```

```python
# 1. Modelni aniqlangan qurilmaga (CPU yoki GPU/CUDA) o'tkazamiz
# Bu model ichidagi og'irliklar (weights/bias) va ma'lumotlar bir joyda hisoblanishi uchun shart
model_1 = model_1.to(device)

# 2. O'qitish davrlari (epochs) sonini 1000 ga teng qilib belgilaymiz
epochs = 1000

# 3. O'qitish uchun ishlatiladigan X (kirish) ma'lumotlarini qurilmaga o'tkazamiz
X_train = X_train.to(device)

# 4. Test qilish uchun ishlatiladigan X (kirish) ma'lumotlarini qurilmaga o'tkazamiz
X_test = X_test.to(device)

# 5. O'qitish uchun ishlatiladigan y (haqiqiy maqsadli javoblar) ma'lumotlarini qurilmaga o'tkazamiz
y_train = y_train.to(device)

# 6. Test qilish uchun ishlatiladigan y (haqiqiy maqsadli javoblar) ma'lumotlarini qurilmaga o'tkazamiz
y_test = y_test.to(device)

# 7. 0 dan 999 gacha (jami 1000 marta) aylanadigan o'qitish tsiklini (for loop) boshlaymiz
for epoch in range(epochs):

    ### --- O'QITISH (TRAINING) QISMI --- ###

    # 8. Modelni o'qitish rejimiga o'tkazamiz (metodlarni faollashtiradi, masalan: Dropout, BatchNorm)
    model_1.train()

    # 9. Forward pass: O'qitish ma'lumotlarini modelga berib, modelning bashoratlarini (y_pred) olamiz
    y_pred = model_1(X_train)

    # 10. Xatolikni hisoblaymiz: Model bergan bashoratlar va haqiqiy javoblar o'rtasidagi farqni o'lchaymiz
    loss = loss_fn(y_pred, y_train)

    # 11. Optimizator ichidagi eski gradientlarni (hosilalarni) nollaymiz, aks holda ular bir-biriga qo'shilib ketadi
    optimizer.zero_grad()

    # 12. Backward pass: Xatolikni orqaga qarab hisoblab, har bir parametr uchun xatolik gradientini topamiz
    loss.backward()

    # 13. Optimizator qadami: Hisoblangan gradientlar yordamida model og'irliklarini (weights/bias) yangilaymiz
    optimizer.step()

    ### --- TEKSHIRISH (TESTING) QISMI --- ###

    # 14. Modelni baholash/test rejimiga o'tkazamiz (o'qitish rejimiga xos metodlarni o'chiradi)
    model_1.eval()

    # 15. Blok ochamiz: Test paytida gradientlarni hisoblash shart emas, bu xotira va vaqtni tejaydi
    with torch.inference_mode():

        # 16. Test ma'lumotlarini modelga berib, test uchun bashoratlarni (test_pred) olamiz
        test_pred = model_1(X_test)

        # 17. Test xatoligini hisoblaymiz: Test bashoratlari va haqiqiy test javoblari o'rtasidagi farq
        test_loss = loss_fn(test_pred, y_test)

    # 18. Shart tekshiramiz: Har 100 ta davrda (epoch) natijalarni ekranga chiqaramiz
    if epoch % 100 == 0:

        # 19. Joriy davr, o'qitish xatoligi (Train loss) va test xatoligini (Test loss) konsolga chiqaramiz
        # :.4f yordamida nuqtadan keyin faqat 4 ta raqam ko'rsatilishini ta'minlaymiz
        print(f"Epoch: {epoch} | Train loss: {loss:.4f} | Test loss: {test_loss:.4f}")

    if epoch % 100 == 0:
      print(f"Epoch: {epoch} | Train loss: {loss} | Test loss: {test_loss}")
```

```python
# Modelni baholash/test (evaluation) rejimiga o'tkazamiz
# Bu rejim o'qitish davridagi ba'zi maxsus qatlamlarni (masalan, Dropout, BatchNorm) testga moslab moslashtiradi
model_1.eval()

# Gradientlarni hisoblashni o'chirgan holda test ma'lumotlari ustida bashorat qilamiz
# torch.inference_mode() kodni tezroq ishlashini ta'minlaydi va xotirani tejaydi
with torch.inference_mode():
    # Test ma'lumotlarini (X_test) modelga berib, yakuniy bashoratlarni olamiz
    y_preds = model_1(X_test)

# Olingan bashorat qiymatlarini (tensor) ekranga chiqaramiz
y_preds
```

```python
# Ma'lumotlarni CPU (markaziy protsessor) ga qaytarib, grafikda chizamiz
# (.cpu() qo'shish sababi - matplotlib GPU dagi tensorlar bilan to'g'ridan-to'g'ri ishlay olmaydi)
# Avvalroq yaratilgan plot_predictions funksiyasiga bashoratlarni uzatamiz
plot_predictions(predictions=y_preds.cpu())
```
