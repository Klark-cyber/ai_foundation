```python
missing_data = {
  'cars': ["BMW", None , "Ford","BMW", None, "Ford"],
  'passings': [3, None, 2, None, 6, 9]
}

md = pd.DataFrame(missing_data)

print(md)
```

```python
md.isnull() # boolena qaytaradi.bor datalarga false, null datalarga true qiymat beradi.
md.isnull().sum() #labelga tegishli elementlar orasida null qiymatdagi elementlarning umumiy sonini qaytaradi
```

```python
md.dropna() #default qaysi qatorda xech bolmasa 1 dona null,NaN, None bor bolsa osha qatorni tashlab yuboradi qaysi labelda bolishidan qatiy nazar.
md.dropna(how="all") # faqat qator boylab barcha elementi falsy bolgan qatornigina ochiradi.
md.dropna(axis=1) #bunda qaysi labelda bir dona bolsa ham falsy bolsa osha labelni ochirib yuboradi.
md.dropna(subset=['passings']) #faqat kerakli labeldagi fulsy datalarni olib tashlaydi.qolgan columnlarni etiborsiz qoldiradi
```

```python
md.fillna(0) # tabledagi barcha fulsy kataklarga 0 yoki argument sifatidag kiritilgan textni qoyadi.
md.fillna({"cars": "unknown", "passings": 0}) #cars unknown matni , passingsga 0 qiymat kiritadi
```
